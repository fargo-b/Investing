# Getting Stock Data and plotting the daily adjusted close price

```python
# Plotting share price 'easy'
import plotly.express as px
import pandas as pd
import yfinance as yf
import datetime as dt

# Date
end = dt.datetime.now()
start = dt.datetime(2000,1,1)

# Stock list
stocklist = 'BP.L'

# Getting the data 
data = yf.download(stocklist, start=start, end=end)

# Grabbing the adj close price
Close = data['Adj Close']

# Plotting
fig = px.line(data, y=Close)

fig.update_layout(title=stocklist)

fig.show()
```

```python
# Importing dependencies

import datetime as dt
import pandas as pd
import yfinance as yf
import plotly.offline as pyo
import plotly.graph_objects as go
from plotly.subplots import make_subplots
pyo.init_notebook_mode(connected=True)

# Date Range
end = dt.datetime.now()
start = dt.datetime(2000,1,1)

# Stock list
stocklist = ['ULVR.L', 'BNZL.L', 'GSK.L', 'AZN.L', 'BDEV.L', 'KIE.L', 'PSN.L', 'TW.L', 'TPK.L', 'BBY.L', 'RIO.L', 'ANTO.L', 'AAL.L', 'HOC.L', 'SGE.L', 'CCC.L', 'BP.L', 'CNA.L', 'SSE.L']

# Using Pandas_datareader 
data = yf.download(stocklist, start=start, end=end)
data.head()

# Dataframe inspection
data.index
data.columns

# Grabbing the adj close price
Close = data['Adj Close']

# Descriptive statistics
Close.describe(percentiles = [0.1,0.5,0.9])

# Plotting price graph using plotly
fig = make_subplots(rows=2, cols=1, 
                    vertical_spacing=0.2, subplot_titles=('RIO.L', 'Date'),
                    row_width=[0.2, 0.7])

fig.add_trace(go.Scatter(x=Close.index, y=Close['RIO.L'], name='RIO.L'), row=1, col=1)

# Editing graph
fig.update_layout(
    title = 'Rio Tinto Price Chart',
    xaxis_tickfont_size = 12,
    yaxis = dict(
        title = 'Price (p/Share)',
        titlefont_size = 14,
        tickfont_size = 12
        ),
    autosize = False,
    width = 500,
    height = 500,
    margin = dict(l=50, r=50, b=100,t=100,pad=5),
)
fig.show()
```
